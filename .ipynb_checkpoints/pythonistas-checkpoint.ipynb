### 4.Display data by splitting age in 4 quartiles and labeling the quartiles.

```python
Patients['Converted_dob']=Patients['DateOfBirth'].dt.strftime("%d/%m/%Y")
def age(born):
    born = datetime.strptime(born, "%d/%m/%Y").date()
    today = date.today()
    return today.year - born.year - ((today.month, 
                                      today.day) < (born.month, 
                                                    born.day))
Patients['Age'] = Patients['Converted_dob'].apply(age)
Quartile_Age =Patients['Age'].quantile([ 0.25,  0.5, 0.75,1])
index_Label = (['1st Quantile','2nd Quantile','3rd Quantile','4th Quantile']) 
Quartile_Age.index =index_Label
Quartile_Age
```

### 28.Which reason of visit has maximum mortality rate.


```python
#Join EDVisists and Patients Table
result = pd.merge(EDVisits,Discharges, how="inner", on=["PatientID"])

#get the data of patients who have expired
exp_pa = result[result['DischargeDisposition']=='Expired']

#get the count of patients who have expired for each reason of visit
df=exp_pa.groupby('ReasonForVisit')['DischargeDisposition'].count()

#change series top Dataframe
df=pd.DataFrame(df)
#Reason of visit is changed from index to columns and count of DischargeDisposition is change to patient_count
df.reset_index(inplace=True)
df = df.rename(columns = {'DischargeDisposition':'Patient_count'})

#Determining the percentage of mortility per Reasonfor visit
df['Mortality_rate'] = df['Patient_count']*100/(df['Patient_count']).sum()#.apply(lambda x:100 * x / float(x.sum())).sort_values(ascending= False)
df_sorted=df.sort_values(by='Mortality_rate',ascending = False)


#get the reason for visit with Highest mortality rate
print('Below data shows the reason for visit with the highest mortality rate along with the number of patients who expired')
df_sorted.head(1)
```

### 32.Write a function to print the multipication table of the given number.

```python
def Mult_table_num(gven_numb, k):
    print(gven_numb, " X ", k, " = ", gven_numb * k)
    if (k < 10):
        Mult_table_num(gven_numb, k + 1)
gven_numb = int(input("Enter some Random Number = "))
print("The Multiplication Table of the above given number", gven_numb, ":")
Mult_table_num(gven_numb, 1)
```

### 54.Mortality O.E.

```python
O_cnt = Discharges['PatientID'][Discharges['DischargeDisposition']=='Expired'].count() #Count of Expired patients

D_cnt = Discharges['DischargeDisposition'].count()#Count of Discharged Patients
#D_cnt#Count of Discharged Patients

#Observed Mortality
Obs_mort = round(O_cnt/D_cnt,4)

#Expected Mortality
Exp_mort = Discharges['ExpectedMortality'].mean()

#Mortality O:E
O_E = round(Obs_mort/Exp_mort,3)
print("The Mortality O:E is",O_E)
```
